```python
import pandas as pd
import matplotlib.pyplot as plt

# 노드 좌표 설정
df = pd.DataFrame([(4, 4), (2, 8), (8, 8), (0, 7), (1, 7), (5, 6), (7, 6), (3, 5), (6, 5),
                 (5, 3), (8, 3), (1, 2), (2, 2), (3, 1), (6, 1), (0, 0), (7, 0)], columns=['x', 'y'])
depot = df.iloc[0]

# 경로 설정
routes1 = [0, 13, 15, 11, 12, 0]
routes2 = [0, 10, 16, 14, 9, 0]
routes3 = [0, 7, 1, 4, 3, 0]
routes4 = [0, 5, 6, 2, 8, 0]
routes = [routes1, routes2, routes3, routes4]

# 좌표 찍기
plt.figure()
plt.scatter(df['x'], df['y'], label="Node")
plt.scatter(depot['x'], depot['y'], s=300, marker='*', label="Depot")

# 차량 경로
for i, route in enumerate(routes):
    route_points = df.iloc[route]
    plt.plot(route_points['x'], route_points['y'], linestyle='-', marker='o', linewidth=2, markersize=4, label=f'Vehicle {i+1}')

#plt.legend()
plt.grid(True)
plt.show()
```

```python
import math

# 유클리드 거리 계산 함수
def euclidean_distance(node1, node2):
    return math.sqrt((node1[0] - node2[0]) ** 2 + (node1[1] - node2[1]) ** 2)

# 사용자 입력 예시 (여러 줄 입력)
input_data = """
1 30 40
2 37 52
3 49 49
4 52 64
5 31 62
6 52 33
7 42 41
8 52 41
9 57 58
10 62 42
11 42 57
12 27 68
13 43 67
14 58 48
15 58 27
16 37 69
"""

# 사용자 입력 데이터를 행별로 나누기
lines = input_data.strip().splitlines()

# 각 행에서 번호를 제외하고 x, y 좌표만 추출하여 노드 리스트 만들기
nodes = [tuple(map(int, line.split()[1:])) for line in lines]

# 노드 수
n = len(nodes)

# 거리 리스트 (2D 리스트로 초기화)
distances = [[0] * n for _ in range(n)]

# 거리 계산 및 배열에 저장
for i in range(n):
    for j in range(n):
        if i != j:
            distances[i][j] = int(round(euclidean_distance(nodes[i], nodes[j])))
        else:
            distances[i][j] = 0  # 자기 자신과의 거리는 0

# 결과 리스트 생성: 각 노드마다 거리 값을 콤마로 연결하여 출력
output = []
for i in range(n):
    row = [f"{distances[i][j]}" for j in range(n)]  # 자기 자신도 포함하여 출력
    output.append(", ".join(row))  # 각 행을 콤마로 구분

# 각 행을 엔터로 구분하여 출력
print(",\n".join(output))
```
